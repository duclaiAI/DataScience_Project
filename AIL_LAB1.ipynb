```python
import httpx
from bs4 import BeautifulSoup
import pandas as pd

# URL của trang web
url = "https://www.thegioididong.com/dtdd-apple-iphone"

# Tạo một SSLContext Cho phép legacy renegotiation (Vận dụng các lỗ hổng bảo mật của cấu hình)
ssl_context = httpx.create_ssl_context()
ssl_context.options |= 0x4 # Cho phép. legacy renegotiation

#Gửi response với SSLContext
with httpx.Client(verify=ssl_context) as client:
    response = client.get(url, headers={'User-Agent': 'Mozilla/5.0'})

#-Kiểm tra response

if response.status_code == 200:
  content = response.text

else:
  print(f"Không đọc được trang đích: {response.status_code}")
  exit()

# Đọc content HTML bằng BeautifulSoup
soup = BeautifulSoup(content, 'html.parser')


```

```python
# Tạo danh sách để lưu trữ thông tin sản phẩm
products = []

# Tìm tất cả các sản phẩm trong trang web
product_items = soup.find_all('li', class_='item')

for item in product_items:
    # Lấy thông tin sản phẩm
    product_id = item.get('data-id')
    product_name = item.find('h3').get('title') if item.find('h3') else None
    merged_name = item.find('div', class_='prods-group').get('data-mergename') if item.find('div', class_='prods-group') else None
    price = item.find('strong', class_='price').text if item.find('strong', class_='price') else None
    old_price = item.find('p', class_='price-old').text if item.find('p', class_='price-old') else None

    # Lấy mô tả sản phẩm
    utility = item.find('div', class_='utility')
    description = "\n".join([p.text for p in utility.find_all('p')]) if utility else None

    # Thêm thông tin sản phẩm vào danh sách
    products.append({
        'ID': product_id,
        'Tên sản phẩm': product_name,
        'Tên gộp': merged_name,
        'Giá bán': price,
        'Giá chưa giảm': old_price,
        'Mô tả': description
    })

# Tạo DataFrame từ danh sách sản phẩm
df = pd.DataFrame(products)

# Lưu DataFrame vào file CSV
df.to_csv('san_pham_dien_thoai.csv', index=False, encoding='utf-8-sig')

print("Dữ liệu đã được lưu vào file san_pham_dien_thoai.csv")
```

# Bài tập cào trắc nghiệm Machine Leaning

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.multiplechoicequestions.org/data-science/machine-learning-mcq-questions-and-answers/"
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    print(f"Truy cập thành công trang web")
else:
    print(f"Không thể truy cập trang web với mã lỗi: {response.status_code}")
```

```python
questions = []

# Lấy các thẻ câu hỏi
content = soup.find_all('div', class_='entry-content clear')[0]

# Lấy nội dung câu hỏi
question_contents = [x.get_text(strip = True) for x in content.find_all('h2')]

# Lấy các options câu hỏi
options_list = ['a', 'b', 'c', 'd']
option_contents = []
for option in options_list:
  items = [x.get_text(strip = True) for x in content.find_all('div', class_='optioncontainer') if x.get_text(strip = True)[0] == option]
  items = list(dict.fromkeys(items))
  option_contents.append(items)

# Lấy đáp án đúng và giải thích
answer_items = content.find_all('div', class_='answer')
answer_contents = [x.find('div', class_='optioncontainer').get_text(strip = True)[0] for x in answer_items]
explaination_contents = [x.find('p').get_text(strip = True) for x in answer_items]

# Tạo dataframe
df_mc = pd.DataFrame({'Question': question_contents,
                      'A': option_contents[0],
                      'B': option_contents[1],
                      'C': option_contents[2],
                      'D': option_contents[3],
                      'Answer': answer_contents,
                      'Explaination': explaination_contents})

# Lưu DataFrame vào file CSV
df_mc.to_csv('ML_MultipleChoice.csv', index=False, encoding='utf-8-sig')

print("Dữ liệu đã được lưu vào file ML_MultipleChoice.csv")
```
